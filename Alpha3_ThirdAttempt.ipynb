## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf
from sklearn.datasets import make_moons

#  Import and read the charity_data.csv.
data_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
data_df.head()
```

```python
# make the moon data
X, y = make_moons(n_samples=1500, noise=0.08, random_state=1)

# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df = data_df.drop(columns=["EIN", "NAME"])
application_df.head()
```

```python
# Determine the number of unique values in each column.
application_df.nunique()
```

```python
# Look at APPLICATION_TYPE value counts for binning
counts = application_df["APPLICATION_TYPE"].value_counts()
counts
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
application_types_to_replace = counts[counts < 500].index

replaced_application_df = application_df

# Replace in dataframe
for app in application_types_to_replace:
    replaced_application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
replaced_application_df['APPLICATION_TYPE'].value_counts()
```

```python
replaced_application_df
```

```python
# Look at CLASSIFICATION value counts for binning
class_counts = application_df['CLASSIFICATION'].value_counts()
class_counts
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
class_counts[class_counts > 1]
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
classifications_to_replace = class_counts[class_counts < 1000].index

# Replace in dataframe
for cls in classifications_to_replace:
    replaced_application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure binning was successful
replaced_application_df['CLASSIFICATION'].value_counts()
```

```python
replaced_application_df
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
# Using `drop_first=True` to prevent multicolliniearity
# See https://builtin.com/machine-learning/dummy-variable-trap-pandas
numeric_df = pd.get_dummies(replaced_application_df, drop_first=True)
numeric_df
```

```python
# Split our preprocessed data into our features and target arrays
X = numeric_df.drop('IS_SUCCESSFUL', axis=1).to_numpy()
y = numeric_df["IS_SUCCESSFUL"].to_numpy()

print(X)
print('\n\n---\n\n')
print(y)
```

```python
# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1, stratify=y)

print(X_train)
print('\n\n---\n\n')
print(X_test)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)

print(X_train_scaled)
print('\n\n---\n\n')
print(X_test_scaled)
```

## Compile, Train and Evaluate the Model

```python
from keras import callbacks
# make a function that will create a single sequential model with hyperparameter options
# options include - choices of activation functions between layers, numbers of neurons on layers, numbers of layers

def makeModel(hp): # hyperparameters that will be passed in from a Hyperband tuner variable

  # Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
  number_input_features = len(X_train_scaled[0])
  # layer_1 = 20
  # layer_2 = 10

  # set up the sequential model
  nn = tf.keras.models.Sequential()

  # set up choices for the activation functions to be used in the hidden layers
  activationList = hp.Choice('activation', ['relu', 'tanh', 'sigmoid'])

  # set up the keras tuner to decide (loop) through a number of attributes in the first layer - input_dim = 2 - loop through 1 - 10 neurons on the second layer
  nn.add(
      tf.keras.layers.Dense(
          input_dim=number_input_features,
          activation= activationList,
          units=hp.Int('first_units', min_value=1, max_value=10)
        )
      )

  # set up the keras tuner to test different numbers of hidden layers as well as neurons on those hidden layers
  for num in range(hp.Int('num_layers', 1, 6)):

    # add the layer (or layers) with a neuron (or neurons) on each layer
    nn.add(
        tf.keras.layers.Dense(
            input_dim=number_input_features,
            activation=activationList,
            units=hp.Int(
                'units_' + str(num),
                         min_value=1,
                         max_value=10
                )
        )
    )


  # Output the model
  nn.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))

  # Compile the model
  nn.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

   # Stop early when best accuracy found
  cb = tf.keras.callbacks.EarlyStopping(monitor='accuracy', min_delta=0.01, patience=15, mode='max', baseline=0.72)

  # Train the model
  nn.fit(X_train_scaled, y_train, epochs=25, callbacks=[cb])

  # Check the structure of the model
  nn.summary()

  # return model
  return nn

```

```python
# import keras tuner modules
import keras_tuner as kt
```

```python
# use the Hyperband function in order to run the optimizer function

tunerModel = kt.Hyperband(
    # pass in the function the contains the attributes
    makeModel,
    # pass in the val_accuracy objective function
    objective='val_accuracy',
    # tell the max_epochs
    max_epochs=25,
    # tell the number of hyperband iterations
    hyperband_iterations=2
)
```

The following model takes an hour to run:

```python
# use the .search() from keras tuner to search for the best combo of hyperparameters
tunerModel.search(X_train_scaled, y_train, epochs=25, validation_data=[X_test_scaled, y_test])
```

```python
tunerModel.get_best_hyperparameters(1)[0].values
```

```python
# use get_best_models to extract the hyper parameters and evaluate the model loss and accuracy
bestModel = tunerModel.get_best_models(1)[0]

modelLoss, modelAccuracy = bestModel.evaluate(X_test_scaled, y_test)
```

```python
print('Loss:', modelLoss)
print('Accuracy:', modelAccuracy)
```

```python
# Export our model to HDF5 file
bestModel.save('Optimization1.h5')
```
